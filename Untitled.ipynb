```python
import mlflow
import os
from mlflow.tracking import MlflowClient

# 1. On se connecte au registre local
client = MlflowClient()

# 2. On cherche le dernier run de l'expérience par défaut (ID "0")
# (C'est là que MLflow enregistre si tu n'as pas spécifié d'expérience)
runs = client.search_runs(
    experiment_ids=["1"], 
    max_results=1, 
    order_by=["attribute.start_time DESC"] # Le plus récent en premier
)

if runs:
    last_run = runs[0]
    run_id = last_run.info.run_id
    
    # 3. On construit le chemin physique sur ton Mac
    # artifact_uri ressemble à : file:///Users/allan/...
    artifact_uri = last_run.info.artifact_uri
    clean_path = artifact_uri.replace("file://", "")
    
    # Le dossier complet
    modele_path = os.path.join(clean_path, "model_2_lstm")
    
    print(f"🔎 Analyse du run : {run_id}")
    
    if os.path.exists(modele_path):
        print("\n✅ MODÈLE TROUVÉ !")
        print("---------------------------------------------------")
        print(f"👉 CHEMIN À COPIER DANS MAIN.PY :\n{modele_path}")
        print("---------------------------------------------------")
        
        # Vérification du contenu pour être sûr
        print("\nContenu du dossier :")
        for f in os.listdir(modele_path):
            print(f" - {f}")
    else:
        print(f"\n❌ Le dossier 'model_2_lstm' n'existe pas dans ce run.")
        print(f"Chemin testé : {modele_path}")
        print("As-tu bien exécuté la cellule 'mlflow.tensorflow.log_model' jusqu'au bout ?")

else:
    print("❌ Aucun run trouvé dans MLflow. As-tu lancé l'entraînement ?")
```
